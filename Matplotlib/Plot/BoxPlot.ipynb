Digunakan untuk menunjukkan distribusi data berdasarkan lima angka ringkasan: minimum, kuartil pertama, median, kuartil ketiga, dan maksimum.

```python
data = [np.random.normal(0, std, 100) for std in range(1, 4)]

# vert orientasi box plot
# patch artits utk warna box
plt.boxplot(data, vert=True, patch_artist=True)
plt.title('Box Plot')
plt.xlabel('Group')
plt.ylabel('Value')
plt.show()
```
