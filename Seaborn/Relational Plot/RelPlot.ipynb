```python
import seaborn as sns
```

```python
sns.get_dataset_names()
```

```python
tips=sns.load_dataset('tips')
tips
```

**Plot relasi generik yang dapat digunakan untuk scatter plot dan line plot. Bergantung pada argumen kind, bisa membuat scatter plot atau line plot.**



```python
sns.relplot(x='tip',y='total_bill',data=tips,hue='smoker',kind='line')
```

```python
sns.relplot(x='tip',y='total_bill',data=tips,hue='smoker',kind='scatter')
```
